```python
from litellm import completion
```

```python

response = completion(
    model="ollama/llama3", 
    messages=[{ "content": "respond in 20 words. who are you?","role": "user"}], 
    api_base="http://localhost:11434"
)
print(response)
```

```python
response.choices[0].message.content
```

```python

features = ["Chatbot UI", "Order Details", "Return, Change, Refund", "Recommendation System", "Backend API", "Database", "User Authentication", "Admin Panel", "Payment Gateway", "Notification System", "User Profile", "Product Catalog", "Order Management", "User Management", "Feedback System", "Search", "Filter", "Sort", "Recommendation System", "Analytics", "Reporting", "Security", "Performance", "Scalability", "Documentation", "Testing", "Deployment", "Maintenance", "Support"]

company=  """
Customer : A retail company : XYZ
Problem : The company is facing issues providing relevant details to the customers on their orders. 
Proposed Solution : A chatbot that can provide the details of the orders to the customers and also help them with the return, change, and refund of the orders.
also build a recommendation system to recommend products to the customers based on their previous orders.

Below are the feature in scope for the chatbot:
""" + ", ".join(features)


system_prompt = """You are a Product Manger, you generates the detailed functionality of feature for Customer product

""" + company 






```

```python
system_prompt
```

```python

response = completion(
    model="ollama/llama3", 
    messages=[ { "content": system_prompt,"role": "system"},
        { "content": "Provide the functionality need for Chatbot UI","role": "user"}], 
    api_base="http://localhost:11434"
)
print(response)
```

```python
print(response.choices[0].message.content)
```

```python
target_folder = "../data"

for feature in features:
    response = completion(
        model="ollama/llama3", 
        messages=[ { "content": system_prompt,"role": "system"},
            { "content": f"Provide the functionality need for {feature}","role": "user"}], 
        api_base="http://localhost:11434"
    )
    print(response.choices[0].message.content)
    with open(f"{target_folder}/{feature}.txt", "w") as f:
        f.write(response.choices[0].message.content)
```
